```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
ml_age=np.random.randint(18,45,(100))
py_age=np.random.randint(15,40,(100))
```

```python
print(ml_age)
print(py_age)
```

```python
plt.hist(ml_age)
plt.title("ML age students Histogram")
plt.xlabel("Students age category")
plt.ylabel("NO of student age")
plt.show()
```

```python
bins=[15,20,25,30,35,40,45]
```

```python
plt.hist(ml_age,bins,rwidth=.8,histtype='bar',align='mid')
plt.title("ML age students Histogram")
plt.xlabel("Students age category")
plt.ylabel("NO of student age")
plt.show()
```

```python
plt.hist(ml_age,bins,rwidth=.8,histtype='bar',align='mid',orientation='horizontal')
plt.title("ML age students Histogram")
plt.xlabel("Students age category")
plt.ylabel("NO of student age")
plt.show()
```

```python
plt.figure(figsize=(16,9))
plt.hist([ml_age,py_age],bins,rwidth=.8,histtype='bar',align='mid')
#plt.hist(py_age,bins,rwidth=.8,histtype='bar',align='mid',color='red',label='ml_students',orientation='horizontal')
plt.title("ML age students Histogram")
plt.xlabel("Students age category")
plt.ylabel("NO of student age")
plt.show()
```

```python
from matplotlib import style
style.use('ggplot')
plt.figure(figsize=(16,9))
plt.hist([ml_age,py_age],bins,rwidth=.8,histtype='bar',align='mid')
#plt.hist(py_age,bins,rwidth=.8,histtype='bar',align='mid',color='red',label='ml_students',orientation='horizontal')
plt.title("ML age students Histogram")
plt.xlabel("Students age category")
plt.ylabel("NO of student age")
plt.grid(color='g')
plt.show()
```
