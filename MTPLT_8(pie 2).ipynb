```python
import matplotlib.pyplot as plt
```

```python
classes=['Python','R','MACHINE LEARNING','ARTIFICIAL INTELLIGENCE','DATA SCIENCE']
class1_st=[45,15,35,25,30]
explode=[0.03,0,.2,0,0]
color=['c','b','r','g','y']
text={'fontsize':15}
we={'linewidth':4,'width':1,'edgecolor':'k'}
plt.pie(class1_st,labels=classes,explode=explode,colors=color,autopct='%0.2f%%',shadow=True,radius=1.1
       ,startangle=90,textprops=text,pctdistance=.6,labeldistance=1.4,counterclock=True
       ,wedgeprops=we,center=(4,5),frame=True,rotatelabels=False)
plt.legend(classes)
plt.show()
```
