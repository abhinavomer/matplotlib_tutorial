```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df1=pd.read_csv('googleplaystore.csv')
df1.shape
```

```python
df1=pd.read_csv('googleplaystore.csv',nrows=1000)
df1
```

```python
x=df1['Rating']
y=df1['Reviews']
z=df1['Installs']
```

```python
plt.scatter(x,y)
plt.title("Google play store apps scatter plot")
plt.xlabel('Ratings')
plt.ylabel("Reviews")
```

```python
plt.figure(figsize=(16,9)),
plt.scatter(x,y,c='r',marker='*',s=100,alpha=0.5,linewidth=10,edgecolors='g')
plt.title("Google play store apps scatter plot")
plt.xlabel('Ratings')
plt.ylabel("Reviews")
```

```python
plt.figure(figsize=(16,9)),
plt.scatter(x,y,c='r',marker='*',s=100,alpha=0.5,linewidth=10,edgecolors='g')
plt.scatter(x,z,c='y',marker='o',s=100,alpha=0.5,linewidth=10,edgecolors='c')
plt.title("Google play store apps scatter plot")
plt.xlabel('Ratings')
plt.ylabel("Reviews")
```
