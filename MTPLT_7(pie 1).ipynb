```python
import matplotlib.pyplot as plt
```

```python
classes=['Python','R','MACHINE LEARNING','ARTIFICIAL INTELLIGENCE','DATA SCIENCE']
class1_st=[45,15,35,25,30]
```

```python
plt.pie([2])
plt.show()
```

```python
plt.pie(class1_st)
plt.show()
```

```python
plt.pie(class1_st,labels=classes)
plt.show()
```

```python
classes=['Python','R','MACHINE LEARNING','ARTIFICIAL INTELLIGENCE','DATA SCIENCE']
class1_st=[45,15,35,25,30]
explode=[0.03,0,.2,0,0]
color=['c','b','r','g','y']
text={'fontsize':15}
plt.pie(class1_st,labels=classes,explode=explode,colors=color,autopct='%0.2f%%',shadow=True,radius=1.4
       ,startangle=90,textprops=text)
plt.show()
```
