```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style
```

```python
classes=['Python','R','AI','ML','DS']
class1_s=[30,10,20,25,10]
class2_s=[40,5,20,20,10]
class3_s=[35,5,30,15,15]
```

```python
plt.bar(classes,class1_s)
```

```python
plt.barh(classes,class1_s)
```

```python
plt.bar(classes,class1_s,width=0.5,align='edge',color='red',edgecolor='b',linewidth=5,alpha=0.7,
       linestyle='--',label='Class 1 student',visible='false')
plt.legend()
```

```python
style.use('ggplot')
plt.figure(figsize=(3,4))
plt.bar(classes,class1_s,width=0.5,align='edge',color='k',edgecolor='b',linewidth=5,
       label='Class 1 student',)
plt.legend()
```

```python
style.use('ggplot')
plt.figure(figsize=(16,9))
plt.barh(classes,class1_s,align='edge',color='k',edgecolor='b',linewidth=5,
       label='Class 1 student',)
plt.legend(loc=1)
```

```python
style.use('ggplot')
plt.figure(figsize=(3,4))
plt.bar(classes,class1_s,width=0.5,align='edge',color='k',edgecolor='b',linewidth=5,
       label='Class 1 student',)
plt.title('Bar chart of Class 1 student')
plt.xlabel('Subjects')
plt.ylabel('Students')
plt.legend()
```

```python
style.use('ggplot')
plt.figure(figsize=(3,4))
plt.bar(classes,class1_s,width=0.5,color='k',edgecolor='b',linewidth=5,
       label='Class 1 student')
plt.bar(classes,class2_s,width=0.5,color='m',edgecolor='b',linewidth=5,
       label='Class 2 student')
plt.title('Bar chart of Class 1 student')
plt.xlabel('Subjects')
plt.ylabel('Students')
plt.legend()
```

```python
style.use('ggplot')
classes_index=np.arange(len(classes))
width=0.2
plt.bar(classes_index,class1_s,color='k',label='Class 1 student',width=.2)
plt.bar(classes_index + width,class2_s,color='m',label='Class 2 student',width=.2)
plt.bar(classes_index + width*2,class3_s,color='r',label='Class 3 student',width=.2)
plt.title('Bar chart of Class 1 student')
plt.xticks(classes_index + width,classes,rotation=20)
plt.xlabel('Subjects')
plt.ylabel('Students')
plt.legend()
```
